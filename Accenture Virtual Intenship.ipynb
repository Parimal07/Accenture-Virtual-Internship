### Cleaning second Dataset

### Cleaning Third Dataset

```python
c=pd.read_csv("Content.csv")
c.head()
```

```python
c.drop(columns=['Unnamed: 0','URL','User ID'],axis=1,inplace=True)
```

```python
c.dtypes
```

```python
c.isnull().sum()
```

```python
c.dropna(axis=0,inplace=True)
```

```python
c.isnull().sum()
```

```python
c.rename(columns={'Type':'Content Type'},inplace=True)
c
```

```python
c['Category']=c['Category'].str.replace('"', '').str.lower()
c.head()
```

```python
m=r.merge(c,on='Content ID',how='inner')
m
```

```python
merged_df=m.merge(r_type,on='Reaction Type',how='inner')
merged_df
```

```python
merged_df.dropna(axis=0,inplace=True)
merged_df
```

```python
merged_df=merged_df.reset_index(drop=True)
```

### Cleansed Dataset

```python
merged_df
```

### Top 5 Performing Categories

```python
top_categories=merged_df.groupby('Category')['Score'].sum()
top_categories
top_categories_df = top_categories.reset_index()
top_categories_df.columns = ['Category', 'Total Score']
```

```python
top_categories=top_categories.sort_values(ascending=False)
top_categories.size
```

```python
top_categories
```

```python
merged_df.to_csv("Cleansed Dataset.csv")
```

```python
top_categories.to_csv("Category wise Scores.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_categories_df['Category'].head(5),top_categories_df['Total Score'].head(5), color='skyblue')
plt.xlabel('Category')
plt.ylabel('Category Count')
plt.title('Top 5 Categories')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_categories_df['Total Score'], labels=top_categories_df['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Categories by Total Score')
plt.axis('equal')
plt.show()
```

```python
most_posts=pd.DataFrame()
merged_df['Datetime'] = pd.to_datetime(merged_df['Datetime'], errors='coerce')
most_posts['Month']=merged_df['Datetime'].dt.month
month_counts = most_posts['Month'].value_counts().reset_index()
month_counts.columns = ['Month', 'Post Count']
sorted_months = month_counts.sort_values(by='Post Count', ascending=False)
sorted_months.reset_index(drop=True)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sorted_months['Month'],sorted_months['Post Count'],color='lightpink')
plt.xlabel('Months')
plt.ylabel('Number of Posts')
plt.title('Number of Posts vs Months')
plt.show()
```

```python
top_content=merged_df['Content Type'].value_counts().reset_index()
top_content.rename(columns={'count':'Count'},inplace=True)
top_content
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_content['Content Type'],top_content['Count'],color='lightgreen')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.title('Posts by Content Type')
plt.show()
```
